```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

```

```python
df = pd.read_csv("Covid Live.csv")
df
```

```python
df.fillna(0, inplace=True)  
```

```python
df
```

```python
df.dtypes
```

```python
cols_to_numeric = ['Total\nCases', 'Total\nDeaths', 'Total\nRecovered', 'Population','Total\nTests']
df[cols_to_numeric] = df[cols_to_numeric].apply(pd.to_numeric, errors='coerce')
```

```python
df.dtypes
```

```python
df["CFR"] = df["Total\nDeaths"] / df["Total\nCases"]
df["RecoveryRate"] = df["Total\nRecovered"] / df["Total\nCases"]
df["CasesPerMillion"] = df["Total\nCases"] / df["Population"] * 1e6
df["DeathsPerMillion"] = df["Total\nDeaths"] / df["Population"] * 1e6
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import folium
```

```python
df["Country,\nOther"] = df["Country,\nOther"].str.strip().str.title()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Total\nCases"], bins=50, kde=True)
plt.title("Distribution of Total Cases")
plt.show()
```

```python
sns.scatterplot(x=df["Total\nTests"], y=df["Total\nCases"])
plt.title("Total Tests vs. Total Cases")
plt.show()
```

```python
top_cases = df.nlargest(10, "Total\nCases")
top_deaths = df.nlargest(10, "Total\nDeaths")
print(top_cases)
print(top_deaths)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=top_cases["Country,\nOther"], y=top_cases["Total\nCases"])
plt.xticks(rotation=45)
plt.title("Top 10 Countries by Total Cases")
plt.show()
```
